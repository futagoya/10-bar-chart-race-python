```python
import pandas as pd
import bar_chart_race as bcr
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/futagoya/10-bar-chart-race-python/main/emas_tahunan_rev2.csv')
```

```python
df.head()
```

```python
df = df.set_index('Tahun')
```

```python
df['Acc_Emas'] = df['Acc_Emas'].str.rstrip('%').astype(float)
df['Acc_IHSG'] = df['Acc_IHSG'].str.rstrip('%').astype(float)
df['Acc_LQ45'] = df['Acc_LQ45'].str.rstrip('%').astype(float)
df['Acc_Properti'] = df['Acc_Properti'].str.rstrip('%').astype(float)
df['Acc_Inflasi'] = df['Acc_Inflasi'].str.rstrip('%').astype(float)
df['Acc_Other'] = df['Acc_Other'].str.rstrip('%').astype(float)
```

```python
column_mapping = {'Acc_Emas': 'Emas',
                  'Acc_IHSG': 'IHSG',
                  'Acc_LQ45': 'LQ45',
                  'Acc_Properti': 'Properti',
                  'Acc_Inflasi': 'Inflasi',
                  'Acc_Other': 'Other'}
df = df.rename(columns=column_mapping)
```

```python
bcr.bar_chart_race(df = df, title = "Inflation vs Other Instruments ",period_length=1200, bar_kwargs = {'alpha': 0.9}, orientation='h')
```
